## Описание данных

- user_id — уникальный идентификатор пользователя, совершившего заказ;
- device_type_canonical — тип устройства, с которого был оформлен заказ;
- order_id — уникальный идентификатор заказа;
- order_dt — дата создания заказа;
- order_ts — дата и время создания заказа;
- currency_code — валюта оплаты;
- revenue — выручка от заказа;
- tickets_count — количество купленных билетов;
- days_since_prev — количество дней от предыдущей покупки пользователя, для пользователей с одной покупкой — значение пропущено;
- event_id — уникальный идентификатор мероприятия;
- service_name — название билетного оператора;
- event_type_main — основной тип мероприятия;
- region_name — название региона, в котором прошло мероприятие;
- city_name — название города, в котором прошло мероприятие.

```python
engine = create_engine(connection_string)
```

```python
query = '''
SELECT p.user_id,
       p.device_type_canonical, 
       p.order_id, 
       p.created_dt_msk AS order_dt, 
       p.created_ts_msk AS order_ts, 
       p.currency_code, 
       p.revenue, 
       p.tickets_count, 
       p.created_dt_msk::date - LAG(p.created_dt_msk::date) OVER (PARTITION BY user_id ORDER BY p.created_dt_msk::date) AS days_since_prev,
       e.event_id,
       e.event_name_code AS event_name, 
       p.service_name, 
       e.event_type_main, 
       r.region_name, 
       c.city_name 
FROM afisha.purchases AS p 
LEFT JOIN afisha.events AS e USING (event_id)
LEFT JOIN afisha.city AS c USING (city_id)
LEFT JOIN afisha.regions AS r USING (region_id) 
WHERE (p.device_type_canonical = 'desktop' OR p.device_type_canonical = 'mobile') AND (e.event_type_main != 'фильм')
ORDER BY user_id;
'''
```

```python
df = pd.read_sql_query(query, con=engine)
```

### ***2) Предобработка данных***

```python
df_tenge = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
print (df_tenge)
```

```python
df_tenge.info()
```

```python
df_tenge['data'] = pd.to_datetime(df_tenge['data'])
df_tenge.info()
```

```python
df = pd.merge(df, df_tenge, left_on = 'order_dt', right_on = 'data', how='left')
def into_rub (row):
    if row['currency_code'] == 'kzt':
        return row['revenue'] / row['nominal'] * row['curs']
    else:
        return row['revenue']
df['revenue_rub'] = df.apply(into_rub, axis=1)
df = df.drop(columns=df_tenge.columns)
df.info()
df.head()
```

В исходный датафрейм добавлен новый столбец revenue_rub, отражающий данные об обплате в российских рублях.

```python
def downcast (df):
    float_cols = [c for c in df if df[c].dtype == 'float64']
    int_cols = [c for c in df if df[c].dtype in ['int64', 'int32', 'int16', 'int8']]
    
    for c in float_cols:
        df[c] = pd.to_numeric(df[c], downcast='float')
    for c in int_cols:
        df[c] = pd.to_numeric(df[c], downcast = 'integer')
    return df
df = downcast(df)
df.info()
```

### ***3) Создание профиля пользователя***

```python
df_sorted = new_df.sort_values(['user_id', 'order_ts'])
first_order_df = df_sorted.groupby('user_id').agg({
    'order_ts': 'first',
    'device_type_canonical': 'first',
    'region_name': 'first',
    'service_name': 'first',
    'event_type_main': 'first'
}).reset_index()
first_order_df.rename(columns={
    'order_ts': 'first_order'
}, inplace=True)
last_order_df = df_sorted.groupby('user_id').agg({
    'order_ts': 'last'
}).reset_index()
last_order_df.rename(columns={
    'order_ts': 'last_order'
}, inplace=True)
profile = first_order_df.merge(last_order_df, on='user_id')
total_orders = df.groupby('user_id')['order_id'].count().reset_index()
total_orders.rename(columns={'order_id': 'total_orders'}, inplace=True)
profile = profile.merge(total_orders, on='user_id')
avg_revenue = df.groupby('user_id')['revenue_rub'].mean().reset_index()
avg_revenue.rename(columns={'revenue_rub': 'avg_revenue_rub'}, inplace=True)
profile = profile.merge(avg_revenue, on='user_id')
avg_tickets = df.groupby('user_id')['tickets_count'].mean().round(2).reset_index()
avg_tickets.rename(columns={'tickets_count': 'avg_tickets_count'}, inplace=True)
profile = profile.merge(avg_tickets, on='user_id')
def mean_time(group):
    delta = group['order_ts'].diff().dropna()
    return delta.mean() if len(delta) > 0 else pd.NaT

time_between_orders = df_sorted.groupby('user_id').apply(mean_time).reset_index()
time_between_orders.rename(columns={0: 'avg_time_between_orders'}, inplace=True)
profile = profile.merge(time_between_orders, on='user_id')
profile['first_order'] = pd.to_datetime(profile['first_order'])
profile['last_order'] = pd.to_datetime(profile['last_order'])

profile.head()
profile.info()
```

```python
profile['is_two'] = (profile['total_orders'] >= 2).astype(int)
profile['is_five'] = (profile['total_orders'] >= 5).astype(int)
profile.head()
```

*Был построен новый датафрейм с данными о пользователях. Для этого данные были сгруппированы по пользователям и посчитаны такие показатели, как дата первого заказа, дата последнего заказа, тип устройства, с которого был сделан заказ, регион первого заказа, название сервиса, у которого был приобретен билет, тип мероприятия, количество заказов, средняя выручка, среднее количество билетов в заказе, среднее время между заказами, а также добовлены столбцы is_two — совершил ли пользователь 2 и более заказа и is_five — совершил ли пользователь 5 и более заказов.*

**В ходе предварительного анализа данных были изучены основные статистические показатели, а также удалено определенное количество строк. 
Столбец данных, отражающих количество совершенных пользователем заказов имел максимальным значением 10251 заказ, что явно не может соответствовать действительности и является выбросом. Было принято решение отфильтровать данные по 99 перцентилю для исключения из анализа такого рода данных.

После проведенной фильтрации по количеству билетов был создан обновленный датасет new_profile с исключенными из него выбросами.**

```python
sport_orders = new_profile[new_profile['event_type_main'] == 'спорт']
concert_orders = new_profile[new_profile['event_type_main'] == 'концерты']

total_sport_users = sport_orders['user_id'].nunique()
repeat_sport_users = sport_orders[sport_orders['is_two'] == 1]['user_id'].nunique()
sport_rate = (repeat_sport_users / total_sport_users * 100) if total_sport_users else 0

total_concert_users = concert_orders['user_id'].nunique()
repeat_concert_users = concert_orders[concert_orders['is_two'] == 1]['user_id'].nunique()
concert_rate = (repeat_concert_users / total_concert_users * 100) if total_concert_users else 0

print(f"Процент возврата пользователей, впервые поситивших спорт: {sport_rate:.2f}%")
print(f"Процент возврата пользователей, впервые поситивших концерт: {concert_rate:.2f}%")
```

**Первая гипотеза не нашла подтверждения при проверке. Процент пользователей, впервые поситивших концерт, выше, чем процент тех, кто впервые посетил спортивное мероприятие.**

```python
regional_counts = new_profile.groupby('region_name')['user_id'].nunique().reset_index()
top_regions = regional_counts.sort_values(by='user_id', ascending=False).head(5)
bottom_regions = regional_counts.sort_values(by='user_id', ascending=True).head(5)
top_heart = new_profile[new_profile['region_name'].isin(top_regions['region_name'])]
bottom_heart = new_profile[new_profile['region_name'].isin(bottom_regions['region_name'])]
def calculate_repeat_rate(new_profile):
    total_users = new_profile['user_id'].nunique()
    repeat_users = new_profile[new_profile['is_two'] == 1]['user_id'].nunique()
    return (repeat_users / total_users * 100) if total_users else 0

top_rate = calculate_repeat_rate(top_heart)
bottom_rate = calculate_repeat_rate(bottom_heart)

print(f"Процент повторных заказов в топ-5 регионах: {top_rate:.2f}%")
print(f"Процент повторных заказов в низших регионах: {bottom_rate:.2f}%")
```

**Вторая гипотеза нашла свое подтверждение в ходе проверки. Действительно, пользователи из более активных регионов чаще делают повторный заказ на платформе.**

```python
avg_service = new_profile.groupby('service_name')['avg_revenue_rub'].sum().sort_values()
order_counts = new_profile.groupby('service_name')['total_orders'].count().sort_values()
avg_service_per_order = avg_service / order_counts
print(avg_service_per_order.sort_values())
```

**Большинство данных лежит в диапазоне от 0 до 1500 рублей. Наблюдая данный график, можно сделать вывод о том, что наибольшую среднюю выручку приносили клиенты, делавшие 1 заказ, в то время как пользователи с повторыми заказами приносили меньше средней выручки. Из этого можно сделать следующий вывод: клиенты, возможно, не возвращаются из-за высоких цен на билеты.**

```python
def assign_category(row):
    if row['is_five'] == 1:
        return '5 и более билетов'
    elif row['is_two'] == 1 and row['is_five'] == 0:
        return '2-4 билета'
new_profile['ticket_category'] = new_profile.apply(assign_category, axis=1)
sns.boxplot(x='ticket_category', y='avg_revenue_rub', data=new_profile, order=['2-4 билета', '5 и более билетов'])
plt.xlabel('Категория количества билетов')
plt.ylabel('Средняя выручка')
plt.title('Распределение средней выручки по категориям')
plt.ylim(0, 5000)
plt.show()
```

**В ходе изучения влияния количества билетов на среднюю выручку было установлено следующее: у тех клиентов, которые имеют более 5 билетов, распределение средней выручки смещено в наименьшую сторону, по сравнению с тем, у кого 2 - 4 билета. Данный факт - еще одно подтверждение того, что возвращаются пользователи, кто платит меньше за билеты.**

```python
sns.histplot(new_profile['avg_tickets_count'], bins=10, kde=False)
plt.xlabel('Среднее количество билетов в заказе')
plt.ylabel('Количество')
plt.title('Распределение среднего количества билетов в заказе')
plt.show()
```

**Результаты данного графика легко интерпретировать и объяснить тем, что большинство пользователей покупают билеты на меропрятия в личных нуждах и, как правило, такое количество ограничивается 1-4 билетов за 1 заказ.**

```python
def assign_segment(row):
    if row['avg_tickets_count'] >= 1 and row['avg_tickets_count'] <= 2:
        return '1-2 билета'
    elif row['avg_tickets_count'] >= 2 and row['avg_tickets_count'] <= 3:
        return '2-3 билета'
    elif row['avg_tickets_count'] >= 3 and row['avg_tickets_count'] <= 5:
        return '3-5 билетов'
    elif row['avg_tickets_count'] >= 5:
        return '5 и более билетов'
new_profile['ticket_segment'] = new_profile.apply(assign_segment, axis=1)
segment_counts = new_profile.groupby('ticket_segment')['user_id'].nunique()
repeat_order_stats = new_profile.groupby('ticket_segment')['is_two'].mean()
result = pd.DataFrame({
    'Количество пользователей': segment_counts,
    'Доля повторных заказов': repeat_order_stats
})
print(result)
```

**В ходе проведенной сегментации пользователей было установлено следующее: наибольшее количество пользователей имеет 1-2 билета в заказе (8988), в то время как более 5 билетов в заказе имеют всего 161 пользователь. С долей повторных заказов ситуация следующая: порядка 70 процентов возвращающихся пользователей находятся в сегментах 2-3 билета. А наименьший процент возврата у пользователей в сегменте более 5 билетов.**

```python
new_profile['day_of_week'] = new_profile['first_order'].dt.dayofweek
days_mapping = {
    0: 'Понедельник',
    1: 'Вторник',
    2: 'Среда',
    3: 'Четверг',
    4: 'Пятница',
    5: 'Суббота',
    6: 'Воскресенье'
}
new_profile['day_name'] = new_profile['day_of_week'].map(days_mapping)
days_group = new_profile.groupby('day_name')['user_id'].count().sort_values()
days_repeat_group = new_profile.groupby('day_name')['is_two'].mean().sort_values()
days_group.plot(kind='bar')
plt.show()
```

**Согласно данному графику, день недели не имеет сильного влияния на покупку билета, однако в пятницу и субботу данный показатель чуть выше, чем в остальные дни недели.**

```python
group1 = new_profile[new_profile['total_orders'].between(2, 4)]
group2 = new_profile[new_profile['total_orders'] >= 5]
avg_time_group1 = group1['avg_time_between_orders'].mean()
avg_time_group2 = group2['avg_time_between_orders'].mean()
print(f"Среднее время между заказами для пользователей, совершивших 2-4 заказа: {avg_time_group1}")
print(f"Среднее время между заказами для пользователей, совершивших более 5 заказов: {avg_time_group2}")
```

**Полученная частота покупки билетов объясняется вполне логично: чем больше заказов, тем меньше времени между ними, поскольку анализируется сравнительно небольшой промежуток времени.**

```python
new_profile.info()
```

*В датафреме к данному моменту помимо изначальных данных добавились данные по сегментациям, категориям и дням неделя. Для проведения корреляционного анализа указанные данные также могут пригодится.*

```python
new_profile['days_between'] = new_profile['avg_time_between_orders'].dt.days
correlation_matrix = new_profile[['device_type_canonical', 'region_name', 'service_name', 'event_type_main', 'total_orders', 'days_between', 'day_name', 'ticket_segment']].phik_matrix(interval_cols=['days_between'])
correlation_matrix.loc[correlation_matrix.index != 'total_orders'][['total_orders']].sort_values(by='total_orders', ascending=False)
plt.figure(figsize=(2, 6))
data_heatmap = correlation_matrix.loc[correlation_matrix.index != 'total_orders'][['total_orders']].sort_values(by='total_orders', ascending=False)
sns.heatmap(data_heatmap,
            annot=True,
            fmt='.2f',
            cmap = 'coolwarm', 
            linewidths=0.5,
            cbar=False)
plt.title('Тепловая карта')
plt.show()
```

По данной тепловой карте видно, что некоторые характеристики имеют нулевую корреляцию с количесвтом заказов (название сервиса и тип мероприятия). Для более качественного изучения степени корреляции целесообразнее будет сегментировать total_orders.

```python
plt.hist(new_profile['total_orders'], bins=50)
plt.xlabel('Значения')
plt.ylabel('Частота')
plt.title('Гистограмма распределения значений total_orders')
plt.show()
```

```python
def total_orders_category(row):
    if row['total_orders'] == 1:
        return '1 заказ'
    elif row['total_orders'] >= 2 and row['total_orders'] <= 4:
        return '2-4 заказа'
    elif row['total_orders'] >= 5:
        return 'более 5 заказов'
new_profile['orders_category'] = new_profile.apply(total_orders_category, axis=1)
correlation_matrix = new_profile[['device_type_canonical', 'region_name', 'service_name', 'event_type_main', 'orders_category', 'days_between', 'day_name', 'ticket_segment']].phik_matrix(interval_cols=['days_between'])
correlation_matrix.loc[correlation_matrix.index != 'orders_category'][['orders_category']].sort_values(by='orders_category')
plt.figure(figsize=(2, 6))
data_heatmap = correlation_matrix.loc[correlation_matrix.index != 'orders_category'][['orders_category']].sort_values(by='orders_category')
sns.heatmap(data_heatmap,
            annot=True,
            fmt='.2f',
            cmap = 'coolwarm', 
            linewidths=0.5,
            cbar=False)
plt.title('Тепловая карта')
plt.show()
```

**Сегментация заказов незначительно повлияла на связь с некоторыми категориями, однако в общей тенденции осталась неизменной. Также, если обратиться к первой тепловой карте, можно также наблюдать небольшую зависимость от региона и сегментации билетов в заказе.**
